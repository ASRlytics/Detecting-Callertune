```
from io import BytesIO
from base64 import b64decode
from google.colab import output
from IPython.display import Javascript

RECORD = '''
const sleep  = time => new Promise(resolve => setTimeout(resolve, time))
const b2text = blob => new Promise(resolve => {
  const reader = new FileReader()
  reader.onloadend = e => resolve(e.srcElement.result)
  reader.readAsDataURL(blob)
})
var record = time => new Promise(async resolve => {
  stream = await navigator.mediaDevices.getUserMedia({ audio: true })
  recorder = new MediaRecorder(stream)
  chunks = []
  recorder.ondataavailable = e => chunks.push(e.data)
  recorder.start()
  await sleep(time)
  recorder.onstop = async ()=>{
    blob = new Blob(chunks)
    text = await b2text(blob)
    resolve(text)
  }
  recorder.stop()
})
'''

def record(sec=0):
  print("Speak Now...")
  display(Javascript(RECORD))
  sec += 1
  print('ab shuru hogi', sec)
  while(1):
    s = output.eval_js('record(%d)' % (sec*1000))
    start = time.time()
    print("Done Recording !")
    b = b64decode(s.split(',')[1])
    with open('audio.wav','wb') as f:
      f.write(b)
    segmentation = seg('audio.wav')
    print(segmentation[0][0])
    print(time.time()-start)
    
    
  return b #byte stream
```

As one audio file is approximately more that 15 seconds so we will create its chunks of 3 seconds of all the files and we will train the on image spectogram of those files

```
directory = 'tring_dataset'
for filename in os.listdir(directory):
  if filename.endswith('.wav'):
    fname = os.path.join(directory, filename)
    with contextlib.closing(wave.open(fname,'r')) as f:
      frames = f.getnframes()
      rate = f.getframerate()
      duration = frames / float(rate)
    t1 = 0 * 1000 #Works in milliseconds
    t2 = 3000
    Audio = AudioSegment.from_wav(fname)
    i = 1
    while(t1 < (duration*1000)):
      save_file = 'tring_dataset/split_files/'+filename.split('.')[0]+'_'+str(i)
      if t2 > duration*1000:
        newAudio = Audio[t1:duration]
        newAudio.export(save_file+'.wav', format="wav")
        break
      else:
        newAudio = Audio[t1:t2]
        newAudio.export(save_file+'.wav', format="wav")
        i = i+1
        t1 = t1+3000
        t2 = t2+3000
    print(filename, 'Splitted')
    
```

Now converting all the splitted images to spectograms

```
from pathlib import Path
import matplotlib.pyplot as plt
from scipy.io import wavfile
import os

directory = 'tring_dataset/split_files'

for filename in os.listdir(directory):
  if filename.endswith('.wav'):
    path = os.path.join(directory, filename)
    samplingfrequency, signaldata = wavfile.read(path)
    try:
      pxx, freq, bins, im = plot.specgram(x=signaldata[:,0], Fs=samplingfrequency, noverlap=384, NFFT=512)
    except:
      pxx, freq, bins, im = plot.specgram(x=signaldata, Fs=samplingfrequency, noverlap=384, NFFT=512)
    plot.axis('off')
    plot.savefig("tring_dataset/images_dataset/tring/{}.png".format(filename.split('.')[0]), bbox_inches='tight', dpi=300, frameon='false')
    print(filename, '-> converted to Image') 
```

# So now we will train our Neural Network to classify weather it is a ringtone or human sound

So i will use transfer learning for this as the dataset is very small so we will use less epochs

```
dataset = 'tring_dataset/images_dataset'
```

```
import keras
from keras.preprocessing.image import ImageDataGenerator
```

```
datagen = ImageDataGenerator(
        rescale=1./255,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        validation_split = 0.2)
train_data = datagen.flow_from_directory(
        dataset,
        target_size=(224, 224),
        color_mode = 'rgb',
        batch_size=32,
        class_mode='binary',
        shuffle = True,
        subset = 'training')
valid_data = datagen.flow_from_directory(
        dataset,
        target_size=(224,224),
        color_mode = 'rgb',
        batch_size=32,
        class_mode='binary',
        shuffle = True,
        subset = 'validation')
```

```
vgg_16 = keras.applications.vgg16.VGG16()
```

```
vgg_16.summary()
```

```
from keras.models import Sequential
from keras.layers import  Conv2D, MaxPool2D
from keras.layers import Activation, Dropout, Flatten, Dense
```

```
model = Sequential()
for layer in vgg_16.layers[:-4]:
  model.add(layer)
# model.summary()
```

```
for layer in model.layers:
  layer.trainable = False
```

```
model.add(Conv2D(64, (2, 2)))
model.add(Conv2D(64, (1, 1)))
model.add(Activation('relu'))
model.add(MaxPool2D(pool_size=(2, 2)))
```

```
model.add(Conv2D(64, (2, 2)))
model.add(Activation('relu'))
model.add(MaxPool2D(pool_size=(2, 2)))
```

```
model.add(Flatten()) 
```

```
model.add(Dense(64))
model.add(Activation('relu'))
model.add(Dense(32))
model.add(Activation('relu'))
model.add(Dropout(0.25))
model.add(Dense(1, activation='sigmoid'))
# model.add(Activation('sigmoid'))
```

```
model.summary()
```

```
from keras import metrics
model.compile(loss='binary_crossentropy',
              optimizer='Adam',
              metrics=[metrics.binary_accuracy])
```

```
model.fit_generator(
        train_data,
        epochs=10,
        validation_data=valid_data,
        validation_steps=40)
```

```
model.save('model.h5')
```
